```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
T = np.linspace(300,500,200)

def l(T):
    a = 1.5*10**(-10)
    Kb = 1.38*10**(-23)
    sigma = np.pi*a**2
    P = 8.7*10**(9-(3110/T))
    n = P/(Kb*T)
    num = n*sigma*np.sqrt(2)
    return 1/num

plt.plot(T,l(T))
plt.grid()
plt.xlabel('T[K]')
plt.ylabel(r'$\bar{l}$[m])')
```
