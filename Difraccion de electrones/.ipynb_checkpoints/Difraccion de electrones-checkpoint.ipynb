```python
#Librerias
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import sympy as sym
from scipy import constants as C
```

```python
#Obtenci贸n de los datos
def Data(l:str):
    direccion = l
    data = pd.read_csv(direccion, delimiter='\t')
    return data

Data1 = Data(r"C:\Users\clauw\Documents\Programming\Laboratorio-Intermedio---Pr-cticas20242\Difraccion de electrones\Datos1.txt")
Data2 = Data(r"C:\Users\clauw\Documents\Programming\Laboratorio-Intermedio---Pr-cticas20242\Difraccion de electrones\Datos2.txt")
Data3 = Data(r"C:\Users\clauw\Documents\Programming\Laboratorio-Intermedio---Pr-cticas20242\Difraccion de electrones\Datos1.txt")
Data1
```

```python
V = Data1['U']*1000
V2 = Data2['U']*1000
V3 = Data3['U']*1000
Vg = []
Vg2 = []
Vg3 = []
for i in V:
    Vg.append(i**(-1/2))

for i in V2:
    Vg2.append(i**(-1/2))

for i in V3:
    Vg3.append(i**(-1/2))

D1_1 = Data1['d1']*0.01
D2_1 = Data1['d2']*0.01
D1_2 = Data2['d1']*0.01
D2_2 = Data2['d2']*0.01
D1_3 = Data3['d1']*0.01
D2_3 = Data3['d2']*0.01
```

```python
#Obtener medidas de incertidumbre

#El r^2 de excel
def GetR2(x,y):
    matrix = np.corrcoef(x,y)
    R2 = (matrix[0,1])**2
    return R2

#Residuales 
def Residuals(y,ym):
    e = []
    for i in range(len(y)):
        e.append(y[i]-ym[i])
    return e

#Regresi贸n lineal
param1_1,cov1_1 = np.polyfit(Vg, D1_1, 1,cov=True)
_xc = np.linspace(0,max(Vg),len(Vg))
m1_1 = param1_1[0]
l1_1 = param1_1[1] 
ym1_1 = m1_1*_xc + l1_1

R2c1_1 = GetR2(Vg, D1_1)
resc1_1 = Residuals(D1_1,ym1_1)
error1_1 = np.sqrt(np.diag(cov1_1))

param2_1,cov2_1 = np.polyfit(Vg, D2_1, 1,cov=True)
m2_1 = param2_1[0]
l2_1 = param2_1[1] 
ym2_1 = m2_1*_xc + l2_1

R2c2_1 = GetR2(Vg, D2_1)
resc2_1 = Residuals(D2_1,ym2_1)
error2_1 = np.sqrt(np.diag(cov2_1))


#Regresi贸n lineal
param1_2,cov1_2 = np.polyfit(Vg2, D1_2, 1,cov=True)
_xc2 = np.linspace(0,max(Vg2),len(Vg2))
m1_2 = param1_2[0]
l1_2 = param1_2[1] 
ym1_2 = m1_2*_xc2 + l1_2

R2c1_2 = GetR2(Vg2, D1_2)
resc1_2 = Residuals(D1_2,ym1_2)
error1_2 = np.sqrt(np.diag(cov1_2))

param2_2,cov2_2 = np.polyfit(Vg2, D2_2, 1,cov=True)
m2_2 = param2_2[0]
l2_2 = param2_2[1] 
ym2_2 = m2_2*_xc2 + l2_2

R2c2_2 = GetR2(Vg2, D2_2)
resc2_2 = Residuals(D2_2,ym2_2)
error2_2 = np.sqrt(np.diag(cov2_2))


#Regresi贸n lineal
param1_3,cov1_3 = np.polyfit(Vg3, D1_3, 1,cov=True)
_xc3 = np.linspace(0,max(Vg3),len(Vg3))
m1_3 = param1_3[0]
l1_3 = param1_3[1] 
ym1_3 = m1_3*_xc3 + l1_3

R2c1_3 = GetR2(Vg3, D1_3)
resc1_3 = Residuals(D1_3,ym1_3)
error1_3 = np.sqrt(np.diag(cov1_3))

param2_3,cov2_3 = np.polyfit(Vg3, D2_3, 1,cov=True)
m2_3 = param2_3[0]
l2_3 = param2_3[1] 
ym2_3 = m2_3*_xc3 + l2_3

R2c2_3 = GetR2(Vg3, D2_3)
resc2_3 = Residuals(D2_3,ym2_3)
error2_3 = np.sqrt(np.diag(cov2_3))
error2_3
```

```python
# #Gr谩fico con residuos
fig, axis = plt.subplots(2,3, figsize = (20, 10), gridspec_kw = {"height_ratios":[2, 1]})
ax1 = axis[0,0]
ax2 = axis[1,0]
ax3 = axis[0,1]
ax4 = axis[1,1]
ax5 = axis[0,2]
ax6 = axis[1,2]

fig.suptitle(r'Relaci贸n para la longitud de onda de Bragg con los di谩metros observados',fontsize=14)

ax1.errorbar(Vg, D1_1,color='purple',fmt='.',capsize=0.5,yerr=error1_1[1])
ax1.errorbar(Vg, D2_1,color='orange',fmt='.',capsize=0.5,yerr=error2_1[1])
ax1.plot(_xc,ym1_1,'--',color='purple', label=r'Pendiente para $D_1$: {} $\pm$ {}'.format(round(m1_1,2) , round(error1_1[0],2)))
ax1.plot(_xc,ym2_1,'--',color='orange', label=r'Pendiente para $D_1$: {} $\pm$ {}'.format(round(m1_1,2) , round(error2_1[0],2)))
ax1.set_ylabel(r"Distancia (m)",fontsize=14)
ax1.set_xlabel(r'V^{-1/2} (V)',fontsize=14)
ax1.grid()
ax1.legend()
ax1.tick_params(axis='both',labelsize=13)
ax1.legend(fontsize=14)

xresc = np.linspace(-0.02,0.06,len(Vg))
ax2.scatter(resc1_1,xresc,label = 'Residuos',color='purple')
ax2.scatter(resc2_1,xresc,color='orange')
ax2.plot(xresc,np.zeros_like(xresc),'--',color='k')
ax2.grid()
ax2.legend()
ax2.tick_params(axis='both',labelsize=13)
ax2.legend(fontsize=14)


ax3.errorbar(Vg2, D1_2,color='purple',fmt='.',capsize=0.5,yerr=error1_2[1])
ax3.errorbar(Vg2, D2_2,color='orange',fmt='.',capsize=0.5,yerr=error2_2[1])
ax3.plot(_xc2,ym1_2,'--',color='purple', label=r'Pendiente para $D_1$: {} $\pm$ {}'.format(round(m1_2,2) , round(error1_2[0],2)))
ax3.plot(_xc2,ym2_2,'--',color='orange', label=r'Pendiente para $D_2$: {} $\pm$ {}'.format(round(m2_2,2), round(error2_2[0],2)))
ax3.set_ylabel(r"Distancia (m)",fontsize=14)
ax3.set_xlabel(r'V^{-1/2} (V)',fontsize=14)
ax3.grid()
ax3.legend()
ax3.tick_params(axis='both',labelsize=13)
ax3.legend(fontsize=14)

xresc = np.linspace(-0.02,0.06,len(Vg2))
ax4.scatter(resc1_2,xresc,label = 'Residuos',color='purple')
ax4.scatter(resc2_2,xresc,color='orange')
ax4.plot(xresc,np.zeros_like(xresc),'--',color='k')
ax4.grid()
ax4.legend()
ax4.tick_params(axis='both',labelsize=13)
ax4.legend(fontsize=14)


ax5.errorbar(Vg3, D1_3,color='purple',fmt='.',capsize=0.5,yerr=error1_3[1])
ax5.errorbar(Vg3, D2_3,color='orange',fmt='.',capsize=0.5,yerr=error2_3[1])
ax5.plot(_xc3,ym1_3,'--',color='purple', label=r'Pendiente para $D_1$: {} $\pm$ {}'.format(round(m1_3,2) , round(error1_3[0],2)))
ax5.plot(_xc3,ym2_3,'--',color='orange', label=r'Pendiente para $D_1$: {} $\pm$ {}'.format(round(m1_3,2) , round(error2_3[0],2)))
ax5.set_ylabel(r"Distancia (m)",fontsize=14)
ax5.set_xlabel(r'V^{-1/2} (V)',fontsize=14)
ax5.grid()
ax5.legend()
ax5.tick_params(axis='both',labelsize=13)
ax5.legend(fontsize=14)

xresc = np.linspace(-0.02,0.06,len(Vg3))
ax6.scatter(resc1_3,xresc,label = 'Residuos',color='purple')
ax6.scatter(resc2_3,xresc,color='orange')
ax6.plot(xresc,np.zeros_like(xresc),'--',color='k')
ax6.grid()
ax6.legend()
ax6.tick_params(axis='both',labelsize=13)
ax6.legend(fontsize=14)

fig.tight_layout()
# plt.savefig('Trabajoenclase.pdf')
```
